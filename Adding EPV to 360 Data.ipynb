```python
# Assuming df_match contains match data with match IDs
match_ids = df_match["match_id"].unique()

# Initialize a list to store the event data for each match
all_event_data = []

for match_id in match_ids:  # Loop over each match ID
    # Load event data for the match
    event_data = parser.event(match_id)[0]
    
    # Append the event data for this match to the list
    all_event_data.append(event_data)

# After the loop, concatenate all event data into a single DataFrame
# Assuming event_data for each match is a DataFrame
final_event_data = pd.concat(all_event_data, ignore_index=True)
```

```python
# List of required columns
required_columns = ["match_id", "id", "index", "period", "minute", "second", "team_name", "player_name", 
                    "position_name", "type_name", "outcome_name", "sub_type_name", "under_pressure", 
                    "pass_shot_assist", "pass_goal_assist", "pass_cross", "pass_cut_back", "pass_switch", 
                    "aerial_won", "pass_recipient_name", "pass_length", "pass_angle", "body_part_name", 
                    "play_pattern_name", "x", "y", "end_x", "end_y", "visible_area", "freeze_frame", "event_uuid"]

# Initialize an empty DataFrame for storing concatenated passes
passes_df = pd.DataFrame()

# Iterate over the matches
for match_id in match_ids:
    
    # Iterate over the periods
    for period in [1, 2]:
        # Filter the data for the specific match, period, and passes with null outcomes
        mask_all_teams_passes = (final_merged_data["match_id"] == match_id) & \
                                (final_merged_data["period"] == period) & \
                                (final_merged_data["type_name"] == 'Pass') & \
                                (final_merged_data["outcome_name"].isnull()) & \
                                ~final_merged_data['sub_type_name'].isin(['Goal Kick', 'Throw-in', 'Free Kick', 'Corner'])
    
        # Filter only the existing columns
        filtered_data = final_merged_data.loc[mask_all_teams_passes, 
                                              [col for col in required_columns if col in final_merged_data.columns]]
    
        # Concatenate the filtered data directly into the passes_df DataFrame
        passes_df = pd.concat([passes_df, filtered_data], ignore_index=True)

# Inspect the final passes DataFrame
passes_df
```

# Line-breaking passes

```python
# Define progressive pass calculation based on distance to goal
def is_progressive_pass(x_start, y_start, x_end, y_end, goal_x=120, goal_y=40):
    """Check if the pass is progressive, meaning it reduces the distance to the goal by more than 10%."""
    beginning = np.sqrt(np.square(goal_x - x_start) + np.square(goal_y - y_start))
    end = np.sqrt(np.square(goal_x - x_end) + np.square(goal_y - y_end))
    return (end / beginning) < 0.90  # At least 10% closer

def is_first_defensive_line_broken(pass_start, pass_end, defenders, horizontal_threshold=5):
    """Check if the pass breaks the first defensive line, formed by the two defenders closest horizontally to each other."""
    
    # Get defender positions (ignoring teammates)
    defenders_x = [d['location'][0] for d in defenders if not d['teammate']]
    defenders_y = [d['location'][1] for d in defenders if not d['teammate']]

    # Check if there are enough defenders to form a line
    if len(defenders_x) < 2:
        return False
    
    # Calculate horizontal distances (x-axis) between each defender and the ball carrier (pass_start)
    defender_distances = np.abs(np.array(defenders_x) - pass_start[0])
    
    # Sort defenders by their horizontal proximity to the ball carrier
    sorted_defenders_idx = defender_distances.argsort()
    
    # Loop through sorted defenders and find the first pair that are close enough horizontally
    for i in range(len(sorted_defenders_idx) - 1):
        d1_idx = sorted_defenders_idx[i]
        d2_idx = sorted_defenders_idx[i + 1]
        
        d1_x, d1_y = defenders_x[d1_idx], defenders_y[d1_idx]
        d2_x, d2_y = defenders_x[d2_idx], defenders_y[d2_idx]
        
        # Check if these two defenders are close enough horizontally to each other (within the threshold)
        if abs(d1_x - d2_x) < horizontal_threshold:
            # Check if the pass ends beyond the furthest defender (x-axis)
            if pass_end[0] > max(d1_x, d2_x):
                # This pass is considered line-breaking
                return True

    return False

# Iterate over passes in the DataFrame
line_breaking_passes = []
for idx, row in passes_df.iterrows():
    # Get pass start and end coordinates
    pass_start = np.array([row['x'], row['y']])
    pass_end = np.array([row['end_x'], row['end_y']])
    
    # Check if the pass is progressive
    if is_progressive_pass(row['x'], row['y'], row['end_x'], row['end_y']):
        # Get the freeze_frame data (list of player positions) for this pass
        freeze_frame = row['freeze_frame']
        
        if freeze_frame:
            # Check if the pass breaks the first defensive line
            if is_first_defensive_line_broken(pass_start, pass_end, freeze_frame, horizontal_threshold=5):
                # Store the pass as a line-breaking pass
                line_breaking_passes.append(row)

# Convert the line-breaking passes to a DataFrame
line_breaking_passes_df = pd.DataFrame(line_breaking_passes)

# Inspect the DataFrame
print(f"Identified {len(line_breaking_passes_df)} line-breaking passes.")
line_breaking_passes_df
```
