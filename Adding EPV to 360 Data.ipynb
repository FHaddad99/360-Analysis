```python
import numpy as np
import pandas as pd

# Define progressive pass calculation based on distance to goal
def is_progressive_pass(x_start, y_start, x_end, y_end, goal_x=120, goal_y=40):
    """Check if the pass is progressive, meaning it reduces the distance to the goal by more than 10%."""
    beginning = np.sqrt(np.square(goal_x - x_start) + np.square(goal_y - y_start))
    end = np.sqrt(np.square(goal_x - x_end) + np.square(goal_y - y_end))
    return (end / beginning) < 0.90  # At least 10% closer

def group_defensive_line(defenders_x, defenders_y, horizontal_threshold):
    """Group defenders that are within a certain horizontal distance of each other."""
    if len(defenders_x) < 2:
        return []  # Not enough defenders to form a line

    defenders_sorted = sorted(zip(defenders_x, defenders_y), key=lambda x: x[0])  # Sort by x-coordinates
    defensive_lines = []
    current_line = [defenders_sorted[0]]

    # Iterate through defenders and group them by proximity
    for i in range(1, len(defenders_sorted)):
        d1_x = defenders_sorted[i - 1][0]
        d2_x = defenders_sorted[i][0]

        # If the next defender is within the threshold, add them to the current line
        if abs(d2_x - d1_x) <= horizontal_threshold:
            current_line.append(defenders_sorted[i])
        else:
            defensive_lines.append(current_line)  # Save the current line and start a new one
            current_line = [defenders_sorted[i]]

    # Add the last group
    if current_line:
        defensive_lines.append(current_line)

    return defensive_lines

def is_defensive_line_broken(pass_start, pass_end, defenders, horizontal_threshold=5):
    """Check if the pass breaks any defensive line formed by defenders close to each other horizontally."""
    
    # Get defender positions (ignoring teammates)
    defenders_x = [d['location'][0] for d in defenders if not d['teammate']]
    defenders_y = [d['location'][1] for d in defenders if not d['teammate']]

    # Group defenders into lines based on horizontal proximity
    defensive_lines = group_defensive_line(defenders_x, defenders_y, horizontal_threshold)

    # Check if any defensive line is broken by the pass
    for line in defensive_lines:
        line_x_coords = [d[0] for d in line]  # Get x-coordinates of defenders in the line
        
        # Ensure the pass end surpasses the max x-coordinate of the defenders in the line
        if pass_end[0] > max(line_x_coords):
            return True  # The pass broke this defensive line by surpassing the last defender
    
    return False

# Iterate over passes in the DataFrame
line_breaking_passes = []
for idx, row in passes_df.iterrows():
    # Get pass start and end coordinates
    pass_start = np.array([row['x'], row['y']])
    pass_end = np.array([row['end_x'], row['end_y']])
    
    # Check if the pass is progressive
    if is_progressive_pass(row['x'], row['y'], row['end_x'], row['end_y']):
        # Get the freeze_frame data (list of player positions) for this pass
        freeze_frame = row['freeze_frame']
        
        if freeze_frame:
            # Check if the pass breaks a defensive line
            if is_defensive_line_broken(pass_start, pass_end, freeze_frame, horizontal_threshold=5):
                # Store the pass as a line-breaking pass
                line_breaking_passes.append(row)

# Convert the line-breaking passes to a DataFrame
line_breaking_passes_df = pd.DataFrame(line_breaking_passes)

# Inspect the DataFrame
print(f"Identified {len(line_breaking_passes_df)} line-breaking passes.")
line_breaking_passes_df
```
