```python
final_df = pd.merge(left=match_events_df, right=match_360_df, left_on='id', right_on='event_uuid', how='left')
```

```python
final_df.head(25)
```

```python
# Define the extract_coordinate function
def extract_coordinate(location, index):
    if isinstance(location, (list, tuple)) and len(location) > index:
        return location[index]
    else:
        return None  # Return None if the location is invalid or missing

# Safely extract coordinates directly in final_df
final_df.loc[:, 'x_start'] = final_df['location'].apply(lambda loc: extract_coordinate(loc, 0))
final_df.loc[:, 'y_start'] = final_df['location'].apply(lambda loc: extract_coordinate(loc, 1))

final_df.loc[:, 'x_end'] = final_df['pass_end_location'].apply(lambda loc: extract_coordinate(loc, 0))
final_df.loc[:, 'y_end'] = final_df['pass_end_location'].apply(lambda loc: extract_coordinate(loc, 1))
```

```python
# Step 1: Filter the DataFrame based on the 'index' column within the data
filtered_df = final_df[final_df['index'] == 1770]  # Adjust '1770' if necessary based on data type

# Step 2: Select the specific columns you want to keep
selected_columns = ['id', 'index', 'period', 'minute', 'second', 'type', 'team', 'player', 'location', 'pass_end_location', 'x_start', 'y_start', 'x_end', 'y_end', 'pass_outcome', 'visible_area', 'freeze_frame']
result_df = filtered_df[selected_columns]

# Display the resulting DataFrame
result_df
```

```python
# Assuming event_id is already defined

# Initialize the pitch as a full horizontal pitch
p = Pitch(pitch_type='statsbomb')
fig, ax = p.draw(figsize=(12, 8))

# Correct the slicing syntax
df = result_df.iloc[0:1]  # Use .iloc or .loc with square brackets to slice the DataFrame

# Plot the scatter points
p.scatter(x=df['x_start'], y=df['y_start'], ax=ax)

# Plot the lines with a comet effect
p.lines(xstart=df['x_start'], ystart=df['y_start'], xend=df['x_end'], yend=df['y_end'], ax=ax, comet=True)

# Plot freeze_frame positions
for x in df.iloc[0]['freeze_frame']:
    if x['teammate']:
        color = 'blue'
    else:
        color = 'red'
        
    p.scatter(x=x['location'][0], y=x['location'][1], ax=ax, c=color, s=100)

# Access and reshape the visible area directly from the DataFrame
visible_area = np.array(df.iloc[0]['visible_area']).reshape(-1, 2)

# Plot the visible area as a polygon with the correct settings
p.polygon([visible_area], color=(0, 0, 1, 0.3), ax=ax, zorder=2)

# Optional: Add a title if desired
fig.suptitle("Area captured by Statsbomb 360 data", fontsize=30)

# Show the plot
plt.show()
```

```python
# Assuming event_id is already defined

# Initialize the pitch as a full horizontal pitch
p = Pitch(pitch_type='statsbomb')
fig, ax = p.draw(figsize=(12, 8))

# Correct the slicing syntax
df = result_df.iloc[0:1]  # Use .iloc or .loc with square brackets to slice the DataFrame

# Plot the scatter points
p.scatter(x=df['x_start'], y=df['y_start'], ax=ax)

# Plot the lines with a comet effect
p.lines(xstart=df['x_start'], ystart=df['y_start'], xend=df['x_end'], yend=df['y_end'], ax=ax, comet=True)

# Plot freeze_frame positions and extract positions
player_positions = np.array([x['location'] for x in df.iloc[0]['freeze_frame']])
teammates = np.array([x['teammate'] for x in df.iloc[0]['freeze_frame']])

# Separate the player positions into x and y coordinates
x_positions = player_positions[:, 0]
y_positions = player_positions[:, 1]

# Plot the visible area
visible_area = np.array(df.iloc[0]['visible_area']).reshape(-1, 2)
visible_polygon = Polygon(visible_area, closed=True, edgecolor='black', linestyle="--", facecolor='none')

# Add the visible area polygon to the plot
ax.add_patch(visible_polygon)

# Calculate Voronoi regions for both teams
team1, team2 = p.voronoi(x_positions, y_positions, teammates)

# Add the Voronoi regions for team 1
t1 = p.polygon(team1, ax=ax, color='yellow', ec='black', lw=3, alpha=0.4, zorder=2)

# Add the Voronoi regions for team 2
t2 = p.polygon(team2, ax=ax, color='red', ec='black', lw=3, alpha=0.4, zorder=2)

# Plot the players for each team
p.scatter(x_positions[teammates], y_positions[teammates], color='yellow', edgecolors='black', s=600, ax=ax, zorder=4)
p.scatter(x_positions[~teammates], y_positions[~teammates], color='red', edgecolors='black', s=600, ax=ax, zorder=3)

# Clip Voronoi regions to the visible area using the Polygon object
for polygon in t1:
    polygon.set_clip_path(visible_polygon)
for polygon in t2:
    polygon.set_clip_path(visible_polygon)

# Add title
fig.suptitle("Voronoi diagram for both teams (in the visible area)", fontsize=30)

# Show the plot
plt.show()
```
