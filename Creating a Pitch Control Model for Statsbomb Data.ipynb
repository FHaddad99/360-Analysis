```python
def calc_event_velocities(actions_df, maxspeed=13.8, position_tolerance=30, formation_tolerance=15):
    """
    Calculate player velocities for Pass, Ball Receipt, and Shot events based on player positions,
    including all players (both involved and non-involved) using freeze frame data from the current 
    and next events.
    """
    
    # Ensure the DataFrame has the necessary columns
    required_columns = ['x', 'y', 'end_x', 'end_y', 'duration', 'type_name', 'freeze_frame', 'match_id', 'index', 'period']
    for col in required_columns:
        if col not in actions_df.columns:
            raise ValueError(f"Missing required column: {col}")
    
    # Initialize new velocity columns for the involved player (if needed)
    actions_df['vx'], actions_df['vy'], actions_df['speed'] = np.nan, np.nan, np.nan
    
    # Dictionary to store velocities for both involved and non-involved players, keyed by (match_id, event_id)
    event_velocities = {}
    
    # Calculate velocities
    for idx, row in actions_df.iterrows():
        match_id = row['match_id']  # Use match_id for uniqueness
        event_id = row['index']     # Event ID to track velocities
        
        # Use (match_id, event_id) as the key in the dictionary
        event_velocities[(match_id, event_id)] = []
        
        # Only calculate velocities for Pass events
        if row['type_name'] == 'Pass':
            # Check if 'freeze_frame' exists and is not empty
            freeze_frame_data = row.get('freeze_frame', None)
            
            # Parse freeze_frame_data if it's stored as a string
            if isinstance(freeze_frame_data, str):
                try:
                    freeze_frame_data = ast.literal_eval(freeze_frame_data)  # Safely evaluate string to list/dict
                except (ValueError, SyntaxError):
                    continue
            
            if not isinstance(freeze_frame_data, list) or len(freeze_frame_data) == 0:
                continue  # Skip events without valid freeze_frame data
            
            duration = row['duration']  # Duration of the current event
            
            # Ensure duration is valid
            if pd.isna(duration) or duration <= 0:
                continue  # Skip events with invalid or missing duration
            
            # Get the next event in the same period from `actions_df`
            next_event = actions_df[
                (actions_df['match_id'] == row['match_id']) &
                (actions_df['period'] == row['period']) &
                (actions_df['index'] > row['index']) &
                (actions_df['type_name'].isin(['Shot', 'Pass', 'Ball Receipt']))
            ].head(1)
            
            # Check if the next event exists and has freeze frame data
            if next_event.empty:
                continue
            
            next_event = next_event.iloc[0]  # Get the next event
            
            # Access freeze_frame for next_event
            next_freeze_frame_data = next_event.get('freeze_frame', None)
            if isinstance(next_freeze_frame_data, str):
                try:
                    next_freeze_frame_data = ast.literal_eval(next_freeze_frame_data)  # Safely evaluate string
                except (ValueError, SyntaxError):
                    continue
            
            if not isinstance(next_freeze_frame_data, list) or len(next_freeze_frame_data) == 0:
                continue
            
            # Separate positions from freeze frame for velocity tracking
            current_positions = np.array([np.array(p['location']) for p in freeze_frame_data])
            next_positions = np.array([np.array(p['location']) for p in next_freeze_frame_data])
            
            # Track velocities for all players based on their positions
            for i, player in enumerate(freeze_frame_data):
                player_position = np.array(player['location'])
                player_team = player['teammate']
                is_actor = player['actor']  # Check if the player is the actor (involved player)
    
                # Find the matching next position for velocity calculation
                next_position = None
                distances = np.linalg.norm(next_positions - player_position, axis=1)
                min_distance_idx = np.argmin(distances)
                if distances[min_distance_idx] <= position_tolerance:
                    next_position = next_positions[min_distance_idx]

                if next_position is not None:
                    delta_pos = next_position - player_position
                    vx = delta_pos[0] / duration if duration > 0 else 0
                    vy = delta_pos[1] / duration if duration > 0 else 0
                    speed = np.sqrt(vx**2 + vy**2)
        
                    # Filter unrealistic speeds
                    if speed > maxspeed:
                        vx, vy, speed = np.nan, np.nan, np.nan
        
                    # Store these velocities directly in the freeze frame for each player
                    player['vx'] = vx
                    player['vy'] = vy
                    player['speed'] = speed

                    # Store velocities for all players, including the actor (involved player)
                    event_velocities[(match_id, event_id)].append({
                        'player_position': player_position.tolist(),
                        'vx': vx,
                        'vy': vy,
                        'speed': speed,
                        'teammate': player_team,
                        'actor': is_actor  # Mark whether this player is the actor
                    })

                    # If this is the involved player (actor), also store in the main DataFrame
                    if is_actor:
                        actions_df.at[idx, 'vx'] = vx
                        actions_df.at[idx, 'vy'] = vy
                        actions_df.at[idx, 'speed'] = speed

    # Print the modified DataFrame with velocities for the involved player
    print("Involved players' velocities:")
    print(actions_df[['type_name', 'vx', 'vy', 'speed']])

    # Print velocities for all players (including non-involved) for the first event
    if event_velocities:
        first_event_id = list(event_velocities.keys())[0]
        print(f"\nVelocities for all players (including non-involved) in event {first_event_id}:")
        print(event_velocities[first_event_id])
    else:
        print("No velocities were calculated.")
    
    # Return the modified DataFrame and velocities for all players (including the involved player)
    return actions_df, event_velocities
```

```python
def default_model_params(time_to_control_veto=3):
    """
    Returns the default parameters for pitch control calculation, modified for a 120x80 pitch.
    
    Parameters:
    -----------
    time_to_control_veto: If the probability that another team or player can get to the ball and control it 
    is less than 10^-time_to_control_veto, ignore that player.
    
    Returns:
    -----------
    params: Dictionary of parameters required for the model.
    """
    pitch_length_ratio = 120 / 106  # Ratio of new pitch length to old
    pitch_area_ratio = (120 * 80) / (106 * 68)  # Ratio of new pitch area to old

    params = {}
    # Player and ball control parameters
    params['max_player_accel'] = 7.0  # Max player acceleration (unchanged)
    
    # Scale max player speed by pitch length ratio
    params['max_player_speed'] = 5.0 * pitch_length_ratio  # Max player speed in m/s (scaled)
    
    params['reaction_time'] = 0.7  # Reaction time in seconds (unchanged)
    
    params['tti_sigma'] = 0.45  # Std. dev. of sigmoid function for time-to-intercept uncertainty (unchanged)
    
    params['kappa_def'] = 1.0  # Defender advantage in ball control probability (unchanged)
    
    # Scale ball control parameters (lambda_att and lambda_def) by pitch area ratio
    params['lambda_att'] = 4.3 * pitch_area_ratio  # Attacking ball control parameter (scaled)
    params['lambda_def'] = 4.3 * params['kappa_def'] * pitch_area_ratio  # Defending ball control parameter (scaled)
    
    # Goalkeeper's advantage also scaled by the same factor
    params['lambda_gk'] = params['lambda_def'] * 3.0  # Goalkeepers' advantage (scaled)
    
    # Scale average ball speed by pitch length ratio
    params['average_ball_speed'] = 15.0 * pitch_length_ratio  # Average ball speed in m/s (scaled)
    
    # Convergence parameters
    params['model_converge_tol'] = 0.01  # Convergence when PPCF > 0.99 at a location (unchanged)
    
    # Time to control cutoff parameters for ball control (unchanged)
    params['time_to_control_att'] = time_to_control_veto * np.log(10) * (np.sqrt(3) * params['tti_sigma'] / np.pi + 1 / params['lambda_att'])
    params['time_to_control_def'] = time_to_control_veto * np.log(10) * (np.sqrt(3) * params['tti_sigma'] / np.pi + 1 / params['lambda_def'])
    
    return params
```

```python
def initialize_players(event_data, freeze_frame_data, params, pass_team, event_velocities, field_dimen=(120., 80.)):
    """
    Initialize players for pitch control, using pre-calculated velocities from event_velocities.

    Parameters:
    -----------
    event_data : dict
        Data for the current event (involved player).
    freeze_frame_data : list
        Freeze frame data containing player positions and attributes.
    params : dict
        Model parameters for pitch control.
    pass_team : str
        Name of the passing team.
    event_velocities : dict
        Velocities for both involved and non-involved players.
    field_dimen : tuple, optional
        Dimensions of the field (default is 120x80).
        
    Returns:
    --------
    attacking_players : list
        List of Player objects for the attacking team.
    defending_players : list
        List of Player objects for the defending team.
    """
    
    attacking_players = []
    defending_players = []

    # Get the match_id and event_id from event_data
    match_id = event_data['match_id']
    event_id = event_data['index']

    # Get the velocities for this specific event
    non_involved_velocities_for_event = event_velocities.get((match_id, event_id), [])

    # Process each player from freeze frame and assign velocities from event_velocities
    for player_info in freeze_frame_data:
        player_position = np.array(player_info['location'])
        
        # Find the matching velocity data directly from event_velocities
        player_vel = next((vel for vel in non_involved_velocities_for_event if np.allclose(player_position, vel['player_position'], atol=1e-1)), None)

        # If no matching velocity data found, set velocity to zero
        player_velocity = np.array([0.0, 0.0]) if player_vel is None else np.array([player_vel['vx'], player_vel['vy']])

        # Determine if the player is an attacker or defender
        is_attacker = player_info['teammate']  # If player is a teammate of the passer, they are attacking

        # Create player object based on their role
        player_obj = Player(
            position=player_position,  # Directly pass the initialized position
            velocity=player_velocity,  # Directly pass the initialized velocity
            attacker=is_attacker,      # Attacker (True) or Defender (False)
            params=params,
            GKid=player_info['keeper'] # Goalkeeper status
        )
        
        if is_attacker:
            attacking_players.append(player_obj)
        else:
            defending_players.append(player_obj)

    # Add goalkeeper to defending team if missing
    if not any(player.is_gk for player in defending_players):
        defending_keeper_position = np.array([field_dimen[0] - 6, field_dimen[1] / 2])  # In the center of the penalty box
        defending_players.append(Player(position=defending_keeper_position, velocity=np.array([0.0, 0.0]), 
                                        attacker=False, params=params, GKid=True))

    # Add goalkeeper to attacking team if missing
    if not any(player.is_gk for player in attacking_players):
        attacking_keeper_position = np.array([6, field_dimen[1] / 2])  # In the center of the penalty box
        attacking_players.append(Player(position=attacking_keeper_position, velocity=np.array([0.0, 0.0]), 
                                        attacker=True, params=params, GKid=True))

    print(f"Attacking players initialized: {[player.position for player in attacking_players]}")
    print(f"Defending players initialized: {[player.position for player in defending_players]}")

    return attacking_players, defending_players
```

```python
class Player(object):
    def __init__(self, position, velocity, attacker, params, GKid):
        """
        Initialize the Player object with pre-calculated position and velocity.

        Parameters:
        -----------
        position : np.ndarray
            Player's position on the field (x, y).
        velocity : np.ndarray
            Player's velocity (vx, vy).
        attacker : bool
            Whether the player is part of the attacking team (True for attacker, False for defender).
        params : dict
            Model parameters for pitch control.
        GKid : bool
            Flag indicating if the player is a goalkeeper.
        """
        self.is_gk = GKid
        self.attacker = attacker
        self.vmax = params['max_player_speed']  # Max player speed in m/s
        self.reaction_time = params['reaction_time']  # Reaction time in seconds
        self.tti_sigma = params['tti_sigma']  # Standard deviation of sigmoid function for time-to-intercept
        self.lambda_att = params['lambda_att']  # Attacking lambda parameter
        self.lambda_def = params['lambda_gk'] if self.is_gk else params['lambda_def']  # Defensive lambda parameter
        
        self.PPCF = 0.0  # Initialize pitch control contribution
        
        # Use the passed position and velocity directly
        self.position = np.array(position)
        self.velocity = np.array(velocity)

        # Debugging output
        print(f"Player initialized - Position: {self.position}, Velocity: {self.velocity}, Attacker: {self.attacker}")

    def simple_time_to_intercept(self, r_final):
        """
        Calculate time to intercept a target location (r_final) given current position and velocity.

        Assumes the player continues moving at current velocity for 'reaction_time' seconds and
        then moves at full speed (vmax) to the target.
        """
        self.PPCF = 0.0  # Initialize pitch control value (used in pitch control calculations)
        
        # Calculate reaction time position
        r_reaction = self.position + self.velocity * self.reaction_time
        
        # Time to intercept: reaction time + time to cover remaining distance at vmax
        self.time_to_intercept = self.reaction_time + np.linalg.norm(r_final - r_reaction) / self.vmax
        
        #print(f"Player - Time to intercept: {self.time_to_intercept}") 
        
        return self.time_to_intercept
    
    def probability_intercept_ball(self, T):
        """
        Calculate the probability of the player intercepting the ball at time T based on the time-to-intercept.
        """
        f = 1 / (1. + np.exp(-np.pi / np.sqrt(3.0) / self.tti_sigma * (T - self.time_to_intercept)))
        return f
```

```python
def generate_pitch_control_for_event(match_id, event_id, actions_df, params, event_velocities, field_dimen=(120., 80.), n_grid_cells_x=50):
    """
    Generate pitch control map for a given event using pre-calculated velocities.

    Parameters:
    -----------
    match_id : int
        The ID of the match for which to calculate pitch control.
    event_id : int
        The ID of the event for which to calculate pitch control.
    actions_df : DataFrame
        DataFrame containing the event data with player positions and velocities.
    params : dict
        Model parameters for pitch control.
    event_velocities : dict
        Pre-calculated velocities for all players (both involved and non-involved).
    field_dimen : tuple, optional
        Dimensions of the pitch (length, width). Default is (120., 80.).
    n_grid_cells_x : int, optional
        Number of grid cells along the x-axis for calculating pitch control. Default is 50.

    Returns:
    --------
    PPCFa : np.ndarray
        Pitch control values for the attacking team.
    PPCFd : np.ndarray
        Pitch control values for the defending team.
    xgrid : np.ndarray
        Grid of x-coordinates for the pitch.
    ygrid : np.ndarray
        Grid of y-coordinates for the pitch.
    """
    try:
        # 1. Fetch event data for the specific match_id and event_id
        event_data = actions_df[(actions_df['match_id'] == match_id) & (actions_df['index'] == event_id)].iloc[0]
        freeze_frame_data = event_data['freeze_frame']  # Extract freeze frame data

        # 2. Get the team name of the involved player from the event data in actions_df
        pass_team = event_data['team_name']  # The team performing the action
        
        # 3. Initialize players (both attacking and defending) with the updated initialize_players function
        attacking_players, defending_players = initialize_players(
            event_data=event_data,
            freeze_frame_data=freeze_frame_data,
            params=params,
            pass_team=pass_team,
            event_velocities=event_velocities
        )

        # Combine attacking and defending players into one list
        all_players = attacking_players + defending_players

        # 4. Ball start position
        ball_start_pos = np.array([event_data['x'], event_data['y']])

        # 5. Initialize grid and pitch control arrays
        PPCFa = np.zeros((n_grid_cells_x, n_grid_cells_x))  # Pitch control for attacking team
        PPCFd = np.zeros((n_grid_cells_x, n_grid_cells_x))  # Pitch control for defending team
        xgrid = np.linspace(0, field_dimen[0], n_grid_cells_x)
        ygrid = np.linspace(0, field_dimen[1], n_grid_cells_x)

        # 6. Calculate pitch control at each grid point
        for i, y in enumerate(ygrid):
            for j, x in enumerate(xgrid):
                target_position = np.array([x, y])

                # Calculate pitch control for this point on the pitch using the unified players list
                P_att, P_def = calculate_pitch_control_at_target(
                    target_position=target_position,
                    players=all_players,  # Pass unified player list
                    ball_start_pos=ball_start_pos,
                    params=params
                )

                # Print the pitch control values for this grid point
                #print(f"Grid Point ({i},{j}) at position [{x}, {y}]: P_att = {P_att}, P_def = {P_def}")

                # Assign pitch control values to PPCFa and PPCFd arrays
                PPCFa[i, j] = P_att
                PPCFd[i, j] = P_def

        # 7. Return the calculated arrays and grids
        return PPCFa, PPCFd, xgrid, ygrid

    except Exception as e:
        print(f"Error in pitch control calculation: {e}")
        return None
```

```python
def calculate_pitch_control_at_target(target_position, players, ball_start_pos, params):
    """
    Calculate the pitch control for a specific target position on the pitch.
    
    Parameters:
    -----------
    target_position : array
        The (x, y) position on the pitch where pitch control is to be calculated.
    players : list
        List of Player objects containing both attacking and defending players.
    ball_start_pos : array
        The starting position of the ball.
    params : dict
        Model parameters for pitch control.
    
    Returns:
    --------
    P_att : float
        Pitch control value for the attacking team at the target position.
    P_def : float
        Pitch control value for the defending team at the target position.
    """
    # Calculate ball travel time from start position to target position
    if ball_start_pos is None or any(np.isnan(ball_start_pos)):
        ball_travel_time = 0.0
    else:
        ball_travel_time = np.linalg.norm(target_position - ball_start_pos) / params['average_ball_speed']

    # Separate attacking and defending players based on their classification
    attacking_players = [p for p in players if p.attacker]
    defending_players = [p for p in players if not p.attacker]

    # Get arrival time of the nearest attacking and defending players
    tau_min_att = np.nanmin([p.simple_time_to_intercept(target_position) for p in attacking_players])
    tau_min_def = np.nanmin([p.simple_time_to_intercept(target_position) for p in defending_players])

    # Print ball travel time, tau_min_att, and tau_min_def for each grid point
    #print(f"Grid Point at {target_position}: ball_travel_time = {ball_travel_time}, tau_min_att = {tau_min_att}, tau_min_def = {tau_min_def}")

    # Calculate pitch control based on arrival times
    if tau_min_att - max(ball_travel_time, tau_min_def) >= params['time_to_control_def']:
        return 0., 1.  # Full control for defending team
    elif tau_min_def - max(ball_travel_time, tau_min_att) >= params['time_to_control_att']:
        return 1., 0.  # Full control for attacking team
    else:
        P_att = 1 / (1 + np.exp(-(tau_min_def - tau_min_att)))
        P_def = 1 - P_att
        return P_att, P_def
```

```python
def plot_pitch_control_mplsoccer(PPCFa, PPCFd, xgrid, ygrid, match_id, event_id, field_dimen=(120., 80.)):
    """
    Plot the pitch control map using intuitive color scales for the attacking (red) and defending (blue) teams,
    with a combined colorbar showing defending control (blue) at the bottom, white in the center, and attacking control (red) at the top.

    Parameters:
    -----------
    PPCFa: np.ndarray
        Pitch control array for the attacking team.
    PPCFd: np.ndarray
        Pitch control array for the defending team.
    xgrid: np.ndarray
        X grid coordinates of the pitch control map.
    ygrid: np.ndarray
        Y grid coordinates of the pitch control map.
    match_id: int
        The ID of the match.
    event_id: int
        The ID of the event.
    field_dimen: tuple
        Dimensions of the field.
    """
    # Initialize the StatsBomb pitch
    pitch = Pitch(pitch_type='statsbomb', pitch_color='white', line_color='black',
                  pitch_length=field_dimen[0], pitch_width=field_dimen[1])

    # Create the figure and axis with matplotlib
    fig, ax = plt.subplots(figsize=(10, 7))

    # Draw the pitch on the axis
    pitch.draw(ax=ax)

    # Set the contour levels for both teams (ranging from 0 to 1)
    levels = np.linspace(0, 1, 10) 

    # Create a mask for dominant areas
    att_mask = PPCFa >= PPCFd
    def_mask = PPCFd > PPCFa

    # Plot attacking control (P_att) where it dominates (att_mask)
    pcm_att = ax.contourf(xgrid, ygrid, np.where(att_mask, PPCFa, np.nan), levels=levels, cmap='Reds', alpha=0.9)

    # Plot defending control (P_def) where it dominates (def_mask)
    pcm_def = ax.contourf(xgrid, ygrid, np.where(def_mask, PPCFd, np.nan), levels=levels, cmap='Blues', alpha=0.9)

    # Adjust the midpoint by setting the boundaries and colors correctly
    red_colors = plt.get_cmap('Reds')(np.linspace(0, 1, 128))
    blue_colors = plt.get_cmap('Blues')(np.linspace(0, 1, 128))
    white = np.array([[1, 1, 1, 1]])  # White for the neutral zone

    # Combine blue, white, and red colors with equal spacing around the midpoint (0)
    combined_colors = np.vstack((blue_colors[::-1], white, red_colors))
    combined_cmap = mcolors.LinearSegmentedColormap.from_list("combined_cmap", combined_colors)

    # Create a single combined colorbar with better scaling for the white zone
    sm = plt.cm.ScalarMappable(cmap=combined_cmap, norm=plt.Normalize(vmin=-1, vmax=1))
    sm.set_array([])  # Empty array for the colorbar
    cbar = plt.colorbar(sm, ax=ax)

    # Set ticks and labels for the colorbar
    cbar.set_ticks([-1, 0, 1])  # Ticks at the ends and middle
    cbar.ax.set_yticklabels(['1 (Defending Team)', '0 (Neutral)', '1 (Attacking Team)'])  # Labels for the ticks
    cbar.set_label('Control Probability')

    # Add the title to the plot using the match_id and event_id
    plt.title(f"Control Areas for Match {match_id}, Event {event_id}", fontsize=14)

    # Show the plot
    plt.show()
```

```python
def plot_pitch_control_with_players(PPCFa, PPCFd, xgrid, ygrid, event_data, freeze_frame_data, event_velocities, match_id, event_id, field_dimen=(120., 80.)):
    """
    Plot the pitch control map using the StatsBomb pitch from mplsoccer for both attacking and defending teams,
    and add player positions and a line for the event's start and end positions.

    Parameters:
    -----------
    PPCFa: np.ndarray
        Pitch control array for the attacking team.
    PPCFd: np.ndarray
        Pitch control array for the defending team.
    xgrid: np.ndarray
        X grid coordinates of the pitch control map.
    ygrid: np.ndarray
        Y grid coordinates of the pitch control map.
    event_data: dict
        Event data containing x, y, end_x, end_y.
    freeze_frame_data: list
        List of freeze-frame data containing player positions.
    event_velocities: dict
        Dictionary containing velocities keyed by (match_id, event_id).
    match_id: int
        The ID of the match.
    event_id: int
        The ID of the event.
    field_dimen: tuple, optional
        Dimensions of the field (default is 120x80).
    """
    # Initialize the StatsBomb pitch without figsize
    pitch = Pitch(pitch_type='statsbomb', pitch_color='white', line_color='black',
                  pitch_length=field_dimen[0], pitch_width=field_dimen[1])

    # Create the figure and axis
    fig, ax = plt.subplots(figsize=(10, 7))

    # Draw the pitch on the axis
    pitch.draw(ax=ax)

    # Set the title for the plot
    ax.set_title(f"Pitch Control for Cucurella's Euro 2024 Game Winning Assist", fontsize=16)
    #ax.set_title(f"Pitch Control for Match {match_id}, Event {event_id}", fontsize=16)

    # Plot pitch control for attacking and defending teams using Reds and Blues colormaps
    levels = np.linspace(0, 1, 10)  # Use 10 levels for a gradient

    # Create a mask for dominant areas (attacking and defending)
    att_mask = PPCFa >= PPCFd
    def_mask = PPCFd > PPCFa

    # Plot attacking control (P_att) where it dominates (att_mask)
    pcm_att = ax.contourf(xgrid, ygrid, np.where(att_mask, PPCFa, np.nan), levels=levels, cmap='Reds', alpha=0.9)

    # Plot defending control (P_def) where it dominates (def_mask)
    pcm_def = ax.contourf(xgrid, ygrid, np.where(def_mask, PPCFd, np.nan), levels=levels, cmap='Blues', alpha=0.9)

    # Add a single combined colorbar using the same logic as before
    red_colors = plt.get_cmap('Reds')(np.linspace(0, 1, 128))
    blue_colors = plt.get_cmap('Blues')(np.linspace(0, 1, 128))
    white = np.array([[1, 1, 1, 1]])  # White for the neutral zone

    # Combine blue, white, and red colors into one colormap
    combined_colors = np.vstack((blue_colors[::-1], white, red_colors))
    combined_cmap = mcolors.LinearSegmentedColormap.from_list("combined_cmap", combined_colors)

    # Create a colorbar for the combined pitch control map
    sm = plt.cm.ScalarMappable(cmap=combined_cmap, norm=plt.Normalize(vmin=-1, vmax=1))
    sm.set_array([])  # Empty array for the colorbar
    cbar = plt.colorbar(sm, ax=ax)
    cbar.set_ticks([-1, 0, 1])  # Ticks at the ends and middle
    cbar.ax.set_yticklabels(['1 (Defending Team)', '0 (Neutral)', '1 (Attacking Team)'])  # Labels for the ticks
    cbar.set_label('Control Probability')

    # Get velocities for this specific event
    velocities_for_event = event_velocities.get((match_id, event_id), [])

    # Plot player positions from freeze_frame_data
    for player in freeze_frame_data:
        player_position = player['location']
        is_teammate = player['teammate']  # Use teammate flag to distinguish between attackers and defenders

        # Plot attacking players in red, defending players in blue
        color = 'red' if is_teammate else 'blue'  # Red for attacking players, blue for defending players
        
        # Find velocity for this player
        player_vel = next((vel for vel in velocities_for_event if np.allclose(player_position, vel['player_position'], atol=1e-1)), None)
        
        # Plot player positions
        ax.scatter(player_position[0], player_position[1], color=color, s=100, edgecolor='black', zorder=5)
        
        # Plot velocity arrows (if velocity exists)
        if player_vel:
            vx = player_vel['vx']
            vy = player_vel['vy']
            if vx != 0 or vy != 0:
                ax.arrow(player_position[0], player_position[1], vx * 0.5, vy * 0.5, 
                         head_width=1.5, head_length=1.5, fc=color, ec=color, zorder=6)

    # Plot a line for the ball's movement from start (x, y) to end (end_x, end_y)
    pitch.lines(xstart=event_data['x'], 
            ystart=event_data['y'], 
            xend=event_data['end_x'], 
            yend=event_data['end_y'], 
            comet=True,  # Optional: comet effect to fade the trail of the pass
            color='green',  # You can specify the color of the pass line here
            ax=ax, zorder=5,
            alpha=0.5)  # Ensure you're using the same axis (ax) you're drawing the pitch on

    # Show the plot
    plt.show()

# Example usage:
# Assuming PPCFa, PPCFd, xgrid, ygrid have been generated from generate_pitch_control_for_event

match_id = 3943043  # Replace with your match ID
event_id = 3218     # Replace with your event index
freeze_frame_data = actions_df.loc[(actions_df['match_id'] == match_id) & (actions_df['index'] == event_id), 'freeze_frame'].values[0]  # Extract freeze_frame from actions_df
event_data = actions_df.loc[(actions_df['match_id'] == match_id) & (actions_df['index'] == event_id)].iloc[0]  # Event data from actions_df

# Now call the function, assuming event_velocities is the dictionary keyed by (match_id, event_id)
plot_pitch_control_with_players(
    PPCFa=PPCFa, 
    PPCFd=PPCFd, 
    xgrid=xgrid, 
    ygrid=ygrid, 
    event_data=event_data, 
    freeze_frame_data=freeze_frame_data, 
    event_velocities=event_velocities,  # Pass the event_velocities
    match_id=match_id,  # Pass the match_id
    event_id=event_id  # Pass the event_id
)
```
